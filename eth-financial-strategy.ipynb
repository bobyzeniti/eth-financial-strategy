### Pipeline that gets Flashbots MEV recipient and the value

```python
# Connect to web3 via Alchemy node:

w3 = Web3(Web3.HTTPProvider(alchemy_url))
print(w3.is_connected())
```

```python
import requests

flashbots_api = requests.get("https://blocks.flashbots.net/v1/blocks")

#Last 100 blocks:
flashbots_json = flashbots_api.json()

mev_list = []
```

```python
for j in range(0, 1):
    print(f'This is the iteration: {j}')
    block_number = flashbots_json['blocks'][10]['block_number']

    # Get the block info by block number
    latest_block = w3.eth.get_block(block_number)

    latest_block['transactions'][-1].hex()
    hex_string = latest_block['transactions'][-1].hex()

    ## Get transaction
    transaction = w3.eth.get_transaction(hex_string)

    mev = transaction['value']
    validator = transaction['to']
    relay = transaction['from']
    timestamp = datetime.fromtimestamp(latest_block['timestamp'])

    if relay == '0xDAFEA492D9c6733ae3d56b7Ed1ADB60692c98Bc5':
        mev_list.append(mev)
```
